```python
###### select files from folder and list them
files = []

for file in os.listdir('.'):
    if file.endswith(".txt"):
        files.append(file)


###### sort files increasing temperature
def file_temp(file):
      
    matches = re.findall("(\\d+)", file)
    for i in matches:
        value= int(i)
        if value>10:
            return value

files.sort(key=file_temp)
```

```python
### WATERFALL PLOT ###

title = ['angle' , 'Yobs' , 'Ycalc' , 'Yob_calc']
fig, ax1 = plt.subplots(figsize=(30*cm,25*cm))
viridis = colormaps['viridis'].resampled(6).colors

### varaiables for waterfall plot ###
x_offst = 0.4 #degrees
y_offst = 10000 

for n, file in enumerate(files):
    item = pd.read_csv(f'{file}' , sep=r'\s+' , names=title )
    temps = file_temp(file)
    tt = re.split('_' , file)[1]
    ff = int(re.split(r'\.', tt)[0])

    ### shift for each cycle
    current_x_shift = n * x_offst
    current_y_shift = -n * y_offst

    label_obs='$Y_{obs}$' if n==0 else ""

    ax1.plot(item['angle'] + current_x_shift, item['Yobs']  + current_y_shift, "x" , color='grey' , label=label_obs , markersize = 3 , alpha=0.4 )
    ax1.plot(item['angle'] + current_x_shift, item['Ycalc'] + current_y_shift, color=viridis[n] , label= f'{ff} °C'  , linewidth=1  )
    
### scale bar ### 
# black bar to show the offset
ax1.plot([5.6, 6], [3000,3000], color='black', linewidth=2.5)
# add the text near the bar
ax1.text(5.8,4000, "0.4° (2\u03B8)", 
         ha='center', va='bottom', fontsize=10, fontweight='bold')

ax1.set_xlabel(f"2\u03B8 $(^o)$", fontsize=14)
ax1.set_ylabel('Intensity (a.u.)', fontsize = 14)
ax1.set_xlim(5,25)
ax1.set_yticks([])
ax1.annotate('$\lambda$=0.354450 $\AA$' , xy =(0.83,0.96), xycoords='axes fraction', fontsize = 14)

fig.suptitle('Cu26' , fontsize = 20)
fig.legend(loc='lower center' , ncol=9, fontsize=16 , bbox_to_anchor=(0.5, -0.05))
fig.tight_layout()

plt.savefig('Cu26_t_selection.png', dpi=1000, bbox_inches='tight')
plt.show()
```
