```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import pandas as pd

font_files = fm.findSystemFonts(r'C:\Users\giuli\OneDrive\Documenti\Università\Dottorato\Burocrazia\Logo UNIPV\UNIPV-Font\UNIPV - Font')
for font in font_files:
    fm.fontManager.addfont(font)

cm = 1/2.54
```

```python
### Data Load from Rietv.Ref.
tet_1 = np.loadtxt('Cu26_TET_985_TOT_test2.res' , skiprows = 1)
g_1 = np.loadtxt('Cu26_G_985_TOT_test2.res' , skiprows = 1)

#### delete index 
tot_tet_new  =  np.delete(tet_1.T , 0 ,0)
g_new = np.delete(g_1.T , 0 , 0)

### FromArraysToDataFrames
titles = ['T_C' , 'T_K' , 'rwp' , 'par_a' , 'par_b' , 'par_c' , 'vol' , 'beq' , 'wt']

tot_rs  = tot_tet_new.T
tot_g   = g_new.T

df_tet = pd.DataFrame( tot_rs  , columns = titles )
df_g = pd.DataFrame(  tot_g   , columns = titles)
```

```python
##### 2 plots 

fig , axs = plt.subplots(2,1, figsize = ( 20*cm , 25*cm))


######[0]
axa = axs[0].twinx()
axs[0].plot( df_tet.T_C ,  df_tet.par_c , "x" , color = 'orange' , label = '$c_{TET}$')
axs[0].plot( df_tet.T_C , df_tet.par_a*2**0.5, "x"  , color='#440154' , label = '$a_{TET}$')
axa.plot( df_tet.T_C,  df_tet.par_a - df_tet.par_c/(2**0.5),  "x" , color = '#1f9e89' , label = 'da')
axs[0].set_xlabel('$T_{calc} (^{o}C)$')
axs[0].set_ylabel('cell parameters $(\AA)$')
axs[0].set_xlim(20,1000)
axa.set_ylabel('da $(\AA)$', color='#1f9e89')
axa.tick_params('y', colors='#1f9e89')

########[1]
axs[1].plot(df_tet.T_C , df_tet.vol*2 , "." , color = '#90d743' , label = '$Vol$')
axs[1].set_xlabel('$T_{calc} (^{o}C)$')
axs[1].set_ylabel('$V_{norm}$ $(\AA^{3})$')



fig.suptitle('Cu26' , fontsize = 16 )
plt.figlegend( loc = 'upper right' , ncols = 4 , bbox_to_anchor=(0.70, 0.01))

plt.tight_layout()
plt.savefig('Cu26_params.png', dpi=1000, bbox_inches='tight')
plt.show()
```
